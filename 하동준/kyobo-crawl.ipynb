```python
samples = isbns_ll[0][:100]
ll = []
for isbn in tqdm(samples):
    s=Service('/opt/ml/chromedriver')
    driver = webdriver.Chrome(options=options, service=s)
    driver.implicitly_wait(5)
    url = "http://www.kyobobook.co.kr/product/detailViewKor.laf?mallGb=KOR&ejkGb=KOR&barcode=" + str(isbn) + "&orderClick=JAI"
    driver.get(url)
    keywords = ''
    keyword = driver.find_element_by_class_name("book_keyword")
    for i in keyword.find_elements_by_xpath('.//*'):
        keywords += i.text
        keywords += ' '
    ll.append(keywords)
    # print(keywords)
```
