> # 장르소설

```python
big_links = []
for serie in tqdm(series):
    baseurl = 'http://www.yes24.com/24/Category/Display/001001046' + serie + '?FetchSize=40&ParamSortTp=01&PageNumber='
    links = []
    for i in tqdm(range(1,37)):
        url = baseurl + str(2)
        res = requests.get(url)
        soup = BeautifulSoup(res.content, 'html.parser')
        class1 = soup.find('ul',class_='clearfix')
        if res.status_code == 200:
            for li in class1.find_all('li'):
                a = li.find('a')
                if a:
                    if a['href'][0] == '/':
                        # print(a['href'])
                        yes24 = 'http://yes24.com'
                        product_link = yes24 + a['href']
                        links.append(product_link)
    print(len(links))
    big_links.append(links)
len(big_links)

```

> # 하나씩 링크 드가서 isbn 따기

```python
big_isbns = []
for links in tqdm(big_links):
    isbns = []
    for link in tqdm(links):
        res2 = requests.get(link)
        soup2 = BeautifulSoup(res2.content, 'html.parser')
        # class1 = soup2.find('ul',class_='clearfix')
        # if res.status_code == 200:
        #     for li in class1.find_all('li'):
        tag = '#infoset_specific > div.infoSetCont_wrap > div > table > tbody > tr:nth-child(3) > td'
        isbn = soup2.select(tag)
        if isbn:
            c = isbn[0].text
            # print(isbn)
            isbns.append(c)
            # print(c)
    print(len(isbns))
    big_isbns.append(isbns)
print(len(big_isbns))
```
